## Split Dataset

```python
import pandas as pd
from sklearn.model_selection import train_test_split

titanic=pd.read_csv('../resources/titanic_cleaned.csv')
titanic.head()
```

```python
features=titanic.drop('Survived',axis=1)
labels=titanic['Survived']

x_train,x_test,y_train,y_test =train_test_split(features,labels,test_size=0.4,random_state=42)
x_val,x_test,y_val,y_test =train_test_split(x_test,y_test,test_size=0.5,random_state=42)
```

```python
for dataset in [y_train, y_val, y_test]:
    print(round(len(dataset) / len(labels), 2))
```

### Write out all data

```python
x_train.to_csv('../resources/train_features.csv', index=False)
x_val.to_csv('../resources/val_features.csv', index=False)
x_test.to_csv('../resources/test_features.csv', index=False)

y_train.to_csv('../resources/train_labels.csv', index=False)
y_val.to_csv('../resources/val_labels.csv', index=False)
y_test.to_csv('../resources/test_labels.csv', index=False)
```
