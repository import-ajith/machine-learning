### Read data

```python
import joblib
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings('ignore',category=FutureWarning)
warnings.filterwarnings('ignore',category=DeprecationWarning)


tr_features = pd.read_csv('../resources/train_features.csv')
tr_labels = pd.read_csv('../resources/train_labels.csv', header=None)
```

### Hyperparameter tuning 

    * High regularization(λ),Low complexity(C) make more likely to underfit
    * Low regularization(λ),high complexity(C) make more likely to overfit

```python
def print_results(results):
    
    means = results.cv_results_['mean_test_score']
    stds = results.cv_results_['std_test_score']
    for mean,std,params in zip(means,stds,results.cv_results_['params']):
        print('{} (+/- {}) for {}'.format(round(mean,3),round(std*2,3),params))
```

```python
lr=LogisticRegression()
parameter={
    'C':[0.001,0.01,0.1,1,10,100,1000]
}
cv=GridSearchCV(lr,parameter,cv=5) #5 set of validation
cv.fit(tr_features,tr_labels.values.ravel()) #ravel convert column vector to array

print_results(cv)
```

```python
cv.best_estimator_
```

### Write out pickled model

```python
joblib.dump(cv.best_estimator_,'../resources/LR_model.pkl')
```
